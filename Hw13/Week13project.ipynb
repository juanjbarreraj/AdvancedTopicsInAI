## Project Update: Campus Agent Behavior and Interaction Refinement

Since the previous update, the project has continued moving from a basic integrated prototype into a more complete campus agent simulation. The main focus during this stage has been refining how the agents behave, how they move through the campus environment, how the user interacts with them, and how the backend explains agent decisions.

The project is still not at the final submission stage yet, but the core system is now much closer to the final vision: an interactive AI campus simulation where agents move, reason, deviate from schedules, and communicate with the user.

### Progress Since the Last Update

In the previous stage, the main goal was to connect the simulation, frontend, backend, and NLP system into one working platform. Since then, the project has improved the agent system so that each agent has a stronger identity and more meaningful behavior.

The system now supports a larger group of campus agents instead of only a few sample agents. Each agent has information such as an ID, name, student type, housing location, current location, expected location, next destination, personality, and behavioral status. This makes the simulation feel more realistic because agents are no longer just visual dots on a map. They now represent individual student-like agents with different roles and decision patterns.

### Agent Population and Student Types

The agent population has been expanded to represent different types of students on campus. The main categories are:

* Athlete
* COPA student
* Regular student

These categories affect where agents are expected to go during different parts of the day. For example, athletes may be connected to practice locations, COPA students may be connected to performance or rehearsal spaces, and regular students may move between class, study, and social locations.

The system also includes housing information, which helps determine where agents begin or return during sleep, evening, or rest periods. Some agents live on campus, while others are treated as off-campus students. This makes the simulation more realistic because not every agent begins from the same location or follows the same routine.

### Time-Slot and Schedule Improvements

A major part of the current work has been improving the time-driven behavior of the agents. The simulation uses defined time slots throughout the day, and each time slot affects agent behavior.

The system now separates:

* The current simulation time
* The active time slot
* The agent’s expected location
* The agent’s actual location
* The agent’s next destination

This is important because the agents are not only moving randomly. Their movement is connected to a daily structure. For example, an agent may be expected to attend class during one time slot, move toward practice in another, and choose a social location during free time.

The time-slot system also supports manual time changes, which is useful for demonstration. This allows the project to show different behaviors without waiting for the full simulation clock to progress naturally.

### Expected Location vs. Actual Location

One of the most important improvements is the distinction between an agent’s expected location and actual location.

The expected location represents where the agent should be according to the schedule. The actual location represents where the agent currently is in the simulation. These two values are not always the same.

This allows the system to represent uncertainty and autonomy. An agent may be expected to be in class, at practice, or at rehearsal, but the actual location may show that the agent chose to go somewhere else. This makes the simulation more interesting and more realistic because agents are allowed to deviate from expected behavior.

This feature also supports better explanations. When a user asks why an agent is at a certain location, the system can compare where the agent is with where the agent was supposed to be.

### Rebel Agent System

The project now includes a rebel agent system. A rebel agent is an agent whose actual location is different from its expected location.

The system tracks rebel behavior using values such as:

* `expected_location_now`
* `current_location`
* `is_rebel`
* `rebel_reason`

This allows the interface to identify which agents are not following their expected schedule. A Rebel Agents panel has also been added under the map. This panel lists agents who are currently in a different location from where they were expected to be.

This improves the multi-agent simulation because the user can quickly identify agents who are behaving differently. The user can click a rebel agent, select them on the map, view their agent card, and ask why they are there.

### Movement and Map Behavior

The movement system has also been improved. Agents are shown visually on the campus map and move between locations as the simulation progresses.

The goal is for movement to feel connected to the campus layout rather than random or disconnected. The project has worked on reducing teleporting behavior and making movement smoother. The agents now use target locations and movement versions so that movement only triggers when an agent actually needs to move.

This is important because if every agent restarts its animation every time one agent changes, the map becomes confusing. The movement logic is being refined so that the selected agent can move independently when needed, especially when the user gives a command while the simulation is paused.

### User Interaction Improvements

The user interaction system has become more detailed. The user can click on an agent and view that agent’s current state in the agent information panel. The selected agent panel displays information such as:

* Agent name
* Agent ID
* Student type
* Sport, if applicable
* Housing location
* Current location
* Expected location
* Next destination
* Time slot
* Rebel status
* Rebel reason, if applicable

This helps the user understand the agent before asking questions. It also makes the simulation easier to demonstrate because the selected agent’s behavior can be inspected directly.

### Question Panel and Agent Communication

The question panel is now being used as the main communication bridge between the user and the selected agent. The user can ask questions such as:

* “Who are you?”
* “Who are you with?”
* “Where are you going next?”
* “Why are you here?”
* “Can we miss class?”
* “Please go to Village Park next.”

The goal is for the agent to answer based on its actual state, not with a generic response. For example, if the agent is a rebel agent, the answer should explain where the agent was supposed to be and why it deviated. If the user asks who the agent is with, the backend should check which other agents are in the same location.

This makes the agents feel more aware of their environment.

### Conversation State Development

Another improvement being developed is conversation state. Some questions should not be answered with only one response. For example, if the user asks whether an agent can miss class, the agent should not immediately say yes or no. Instead, the agent should ask for a reason if it is currently expected to be in class, practice, or rehearsal.

This creates a more realistic decision process:

1. The user asks if the agent can miss an obligation.
2. The agent checks whether it is currently expected somewhere important.
3. If the decision requires justification, the agent asks the user for a reason.
4. The user gives a reason.
5. The agent decides whether to accept or refuse based on personality, obligation, and risk.

This is an important step because it makes the system more interactive and more stateful. The agent can now treat a follow-up response as part of the same conversation instead of treating every message as a completely new question.

### AI Evidence and Reasoning Display

The project is also continuing to improve how AI evidence is displayed. The goal is for the user to see not only the agent’s answer, but also the reasoning behind it.

The interface is being developed to show evidence such as:

* Predicted intent
* Confidence score
* Attention weights
* Word representation evidence
* Campus AI reasoning
* Expected utility evidence
* Value of information
* Multi-agent congestion
* EM hidden activity-mode inference
* Q-learning action recommendations

This is important for the final project because it shows that the system is not just producing text responses. It is connecting language understanding, agent state, decision-making, and reasoning evidence.

### Current Challenges

The main challenge is keeping all parts of the system synchronized. The frontend map, selected agent panel, time-slot system, question panel, and backend reasoning all need to refer to the same agent state.

Another challenge is making sure agent movement remains smooth and understandable. Since many agents are moving at the same time, the simulation must avoid visual confusion.

A third challenge is making responses specific enough. The agent should not simply say, “I am here because it matches my schedule.” Instead, the answer should explain the actual reason based on the time slot, student type, expected location, current location, personality, and rebel status.

A final challenge is performance. Because the system includes several AI components, the local version is more reliable for full testing. The live deployment is being explored, but backend AI modules may be heavier than what free hosting services handle easily.
