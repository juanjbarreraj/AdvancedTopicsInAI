```python

"""
Module 10 - Neural NLP System (improved)
Runs as a single Python file and renders plots inline with plt.show().

Recommended install:
    pip install torch transformers scikit-learn matplotlib numpy

Notebook usage:
    %matplotlib inline
    %run module10_nlp_system_improved.py
"""

import math
import random
import re
from collections import Counter, defaultdict
from dataclasses import dataclass

import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
    ConfusionMatrixDisplay,
)
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

from transformers import AutoTokenizer, AutoModelForSequenceClassification


# ============================================================
# GLOBAL SETTINGS
# ============================================================

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["axes.grid"] = True

print(f"\nUsing device: {DEVICE}\n")


# ============================================================
# HELPERS
# ============================================================

def simple_tokenize(text: str):
    text = text.lower().strip()
    text = re.sub(r"[^a-z0-9\s']", "", text)
    return text.split()

def plot_curve(values, title, ylabel):
    plt.figure()
    plt.plot(range(1, len(values) + 1), values)
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel(ylabel)
    plt.show()

def compute_classification_metrics(y_true, y_pred):
    precision, recall, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="binary", zero_division=0
    )
    acc = accuracy_score(y_true, y_pred)
    return {
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }


# ============================================================
# DATA GENERATION FOR PARTS 1-3
# ============================================================

def build_synthetic_corpus():
    sports_subjects = ["athlete", "runner", "player", "sprinter", "coach"]
    sports_verbs = ["trains", "runs", "practices", "sprints", "lifts"]
    sports_places = ["gym", "track", "field", "stadium"]
    sports_times = ["morning", "evening", "afternoon"]

    education_subjects = ["student", "teacher", "researcher", "learner", "tutor"]
    education_verbs = ["studies", "reads", "writes", "learns", "reviews"]
    education_places = ["library", "classroom", "lab", "campus"]
    education_times = ["morning", "night", "weekend"]

    arts_subjects = ["musician", "artist", "actor", "painter", "singer"]
    arts_verbs = ["plays", "rehearses", "creates", "paints", "performs"]
    arts_places = ["studio", "stage", "theater", "gallery"]
    arts_times = ["evening", "night", "weekend"]

    adjectives = ["dedicated", "focused", "talented", "disciplined", "creative"]
    frequency = ["every day", "every week", "after class", "before practice"]

    sentences = []
    labeled = []

    for _ in range(220):
        s = random.choice(sports_subjects)
        v = random.choice(sports_verbs)
        p = random.choice(sports_places)
        t = random.choice(sports_times)
        a = random.choice(adjectives)
        f = random.choice(frequency)
        variants = [
            f"the {a} {s} {v} in the {p} every {t}",
            f"the {s} {v} in the {p} {f}",
            f"the {s} {v} with the coach at the {p}",
            f"the {s} {v} hard in the {p} during the {t}",
        ]
        text = random.choice(variants)
        sentences.append(text)
        labeled.append((text, "sports"))

    for _ in range(220):
        s = random.choice(education_subjects)
        v = random.choice(education_verbs)
        p = random.choice(education_places)
        t = random.choice(education_times)
        a = random.choice(adjectives)
        f = random.choice(frequency)
        variants = [
            f"the {a} {s} {v} in the {p} every {t}",
            f"the {s} {v} in the {p} {f}",
            f"the {s} {v} notes with the teacher in the {p}",
            f"the {s} {v} carefully in the {p} during the {t}",
        ]
        text = random.choice(variants)
        sentences.append(text)
        labeled.append((text, "education"))

    for _ in range(220):
        s = random.choice(arts_subjects)
        v = random.choice(arts_verbs)
        p = random.choice(arts_places)
        t = random.choice(arts_times)
        a = random.choice(adjectives)
        f = random.choice(frequency)
        variants = [
            f"the {a} {s} {v} in the {p} every {t}",
            f"the {s} {v} in the {p} {f}",
            f"the {s} {v} with the director at the {p}",
            f"the {s} {v} creatively in the {p} during the {t}",
        ]
        text = random.choice(variants)
        sentences.append(text)
        labeled.append((text, "arts"))

    random.shuffle(sentences)
    random.shuffle(labeled)
    return sentences, labeled

sentences, labeled_sentences = build_synthetic_corpus()


# ============================================================
# PART 1 - WORD REPRESENTATIONS (CBOW + SIMILARITY + PCA)
# ============================================================

print("=" * 70)
print("PART 1 - WORD REPRESENTATIONS WITH EMBEDDINGS")
print("=" * 70)

tokenized_sentences = [simple_tokenize(s) for s in sentences]
all_words = [w for sent in tokenized_sentences for w in sent]
vocab = sorted(set(all_words))
word_to_ix = {w: i for i, w in enumerate(vocab)}
ix_to_word = {i: w for w, i in word_to_ix.items()}
vocab_size = len(vocab)

context_size = 2
training_pairs = []

for sent in tokenized_sentences:
    if len(sent) < 2 * context_size + 1:
        continue
    for i in range(context_size, len(sent) - context_size):
        context = sent[i - context_size:i] + sent[i + 1:i + context_size + 1]
        target = sent[i]
        training_pairs.append((context, target))

train_pairs, val_pairs = train_test_split(training_pairs, test_size=0.2, random_state=SEED)

class CBOW(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super().__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear1 = nn.Linear(embedding_dim, 64)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(64, vocab_size)

    def forward(self, inputs):
        embeds = self.embeddings(inputs)
        pooled = embeds.mean(dim=0).view(1, -1)
        x = self.linear1(pooled)
        x = self.relu(x)
        x = self.linear2(x)
        return x

def evaluate_cbow(model, pairs):
    model.eval()
    loss_fn = nn.CrossEntropyLoss()
    total = 0.0
    with torch.no_grad():
        for context, target in pairs:
            context_idxs = torch.tensor([word_to_ix[w] for w in context], dtype=torch.long).to(DEVICE)
            target_idx = torch.tensor([word_to_ix[target]], dtype=torch.long).to(DEVICE)
            logits = model(context_idxs)
            loss = loss_fn(logits, target_idx)
            total += loss.item()
    return total / max(1, len(pairs))

embedding_dim = 32
cbow_model = CBOW(vocab_size, embedding_dim).to(DEVICE)
cbow_loss_fn = nn.CrossEntropyLoss()
cbow_optimizer = optim.Adam(cbow_model.parameters(), lr=0.005)

cbow_train_losses = []
cbow_val_losses = []

for epoch in range(30):
    cbow_model.train()
    total_loss = 0.0
    random.shuffle(train_pairs)

    for context, target in train_pairs:
        context_idxs = torch.tensor([word_to_ix[w] for w in context], dtype=torch.long).to(DEVICE)
        target_idx = torch.tensor([word_to_ix[target]], dtype=torch.long).to(DEVICE)

        cbow_optimizer.zero_grad()
        logits = cbow_model(context_idxs)
        loss = cbow_loss_fn(logits, target_idx)
        loss.backward()
        nn.utils.clip_grad_norm_(cbow_model.parameters(), 1.0)
        cbow_optimizer.step()
        total_loss += loss.item()

    avg_train = total_loss / len(train_pairs)
    avg_val = evaluate_cbow(cbow_model, val_pairs)
    cbow_train_losses.append(avg_train)
    cbow_val_losses.append(avg_val)

    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch + 1:2d} | CBOW Train Loss = {avg_train:.4f} | Val Loss = {avg_val:.4f}")

def cosine_similarity(vec1, vec2):
    return torch.dot(vec1, vec2) / (torch.norm(vec1) * torch.norm(vec2) + 1e-8)

def most_similar(model, word, top_k=5):
    model.eval()
    if word not in word_to_ix:
        return []
    with torch.no_grad():
        emb = model.embeddings.weight.data
        idx = word_to_ix[word]
        vec = emb[idx]
        sims = []
        for w, j in word_to_ix.items():
            if w == word:
                continue
            sims.append((w, cosine_similarity(vec, emb[j]).item()))
        sims.sort(key=lambda x: x[1], reverse=True)
        return sims[:top_k]

print("\nEmbedding similarity results:")
similarity_results = {}
for probe in ["athlete", "student", "musician", "library", "studio", "track"]:
    sims = most_similar(cbow_model, probe, top_k=5)
    similarity_results[probe] = sims
    print(f"\nMost similar to '{probe}':")
    for neighbor, score in sims:
        print(f"  {neighbor:12s} -> {score:.4f}")

# Plot training and validation loss
plt.figure()
plt.plot(range(1, len(cbow_train_losses) + 1), cbow_train_losses, label="Train")
plt.plot(range(1, len(cbow_val_losses) + 1), cbow_val_losses, label="Validation")
plt.title("CBOW Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Similarity plot for one example word
example_word = "athlete"
labels = [w for w, _ in similarity_results[example_word]]
scores = [s for _, s in similarity_results[example_word]]
plt.figure()
plt.bar(labels, scores)
plt.title(f"Top Similar Words to '{example_word}'")
plt.xlabel("Word")
plt.ylabel("Cosine Similarity")
plt.xticks(rotation=30)
plt.show()

# PCA embedding plot
cbow_model.eval()
with torch.no_grad():
    embedding_matrix = cbow_model.embeddings.weight.detach().cpu().numpy()

probe_words = ["athlete", "runner", "player", "student", "teacher", "researcher", "musician", "artist", "actor",
               "gym", "track", "library", "classroom", "studio", "stage"]
probe_indices = [word_to_ix[w] for w in probe_words]
probe_vectors = embedding_matrix[probe_indices]
pca = PCA(n_components=2, random_state=SEED)
coords = pca.fit_transform(probe_vectors)

plt.figure(figsize=(9, 6))
for (x, y), word in zip(coords, probe_words):
    plt.scatter(x, y)
    plt.text(x + 0.02, y + 0.02, word, fontsize=9)
plt.title("PCA Projection of Selected Word Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()


# ============================================================
# PART 2 - SEQUENCE MODELING (LSTM NEXT-WORD PREDICTION)
# ============================================================

print("\n" + "=" * 70)
print("PART 2 - SEQUENCE MODELING WITH LSTM")
print("=" * 70)

sequence_sentences = sentences[:450]
tokenized_seq = [simple_tokenize(s) for s in sequence_sentences]

seq_vocab = sorted(set(word for sent in tokenized_seq for word in sent))
seq_word_to_ix = {w: i + 1 for i, w in enumerate(seq_vocab)}  # 0 reserved for padding
seq_ix_to_word = {i: w for w, i in seq_word_to_ix.items()}
seq_vocab_size = len(seq_word_to_ix) + 1

X_seq = []
y_seq = []

for sent in tokenized_seq:
    for i in range(1, len(sent)):
        prefix = sent[:i]
        target = sent[i]
        X_seq.append(prefix)
        y_seq.append(target)

max_len_seq = max(len(x) for x in X_seq)

def encode_prefix(seq, mapping, max_len):
    ids = [mapping[w] for w in seq]
    ids += [0] * (max_len - len(ids))
    return ids

X_seq_enc = np.array([encode_prefix(x, seq_word_to_ix, max_len_seq) for x in X_seq], dtype=np.int64)
y_seq_enc = np.array([seq_word_to_ix[y] for y in y_seq], dtype=np.int64)

X_train, X_val, y_train, y_val = train_test_split(
    X_seq_enc, y_seq_enc, test_size=0.2, random_state=SEED
)

X_train_t = torch.tensor(X_train, dtype=torch.long).to(DEVICE)
X_val_t = torch.tensor(X_val, dtype=torch.long).to(DEVICE)
y_train_t = torch.tensor(y_train, dtype=torch.long).to(DEVICE)
y_val_t = torch.tensor(y_val, dtype=torch.long).to(DEVICE)

class NextWordLSTM(nn.Module):
    def __init__(self, vocab_size, emb_dim=48, hidden_dim=96):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.lstm = nn.LSTM(emb_dim, hidden_dim, batch_first=True)
        self.dropout = nn.Dropout(0.15)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x):
        emb = self.embedding(x)
        _, (hidden, _) = self.lstm(emb)
        last_hidden = self.dropout(hidden[-1])
        logits = self.fc(last_hidden)
        return logits

def top_k_accuracy(logits, targets, k=3):
    topk = torch.topk(logits, k=k, dim=1).indices
    correct = (topk == targets.unsqueeze(1)).any(dim=1).float().mean().item()
    return correct

lstm_model = NextWordLSTM(seq_vocab_size).to(DEVICE)
lstm_loss_fn = nn.CrossEntropyLoss()
lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=0.003)

lstm_train_losses, lstm_val_losses = [], []
lstm_train_accs, lstm_val_accs = [], []
lstm_train_top3, lstm_val_top3 = [], []

for epoch in range(20):
    lstm_model.train()
    lstm_optimizer.zero_grad()
    train_logits = lstm_model(X_train_t)
    train_loss = lstm_loss_fn(train_logits, y_train_t)
    train_loss.backward()
    nn.utils.clip_grad_norm_(lstm_model.parameters(), 1.0)
    lstm_optimizer.step()

    lstm_model.eval()
    with torch.no_grad():
        val_logits = lstm_model(X_val_t)
        val_loss = lstm_loss_fn(val_logits, y_val_t)

        train_preds = torch.argmax(train_logits, dim=1)
        val_preds = torch.argmax(val_logits, dim=1)

        train_acc = (train_preds == y_train_t).float().mean().item()
        val_acc = (val_preds == y_val_t).float().mean().item()

        train_top3 = top_k_accuracy(train_logits, y_train_t, 3)
        val_top3 = top_k_accuracy(val_logits, y_val_t, 3)

    lstm_train_losses.append(train_loss.item())
    lstm_val_losses.append(val_loss.item())
    lstm_train_accs.append(train_acc)
    lstm_val_accs.append(val_acc)
    lstm_train_top3.append(train_top3)
    lstm_val_top3.append(val_top3)

    if (epoch + 1) % 5 == 0:
        print(
            f"Epoch {epoch + 1:2d} | "
            f"Train Loss = {train_loss.item():.4f} | Val Loss = {val_loss.item():.4f} | "
            f"Train Acc = {train_acc:.4f} | Val Acc = {val_acc:.4f} | "
            f"Val Top-3 = {val_top3:.4f}"
        )

def predict_next_word(prefix, top_k=3):
    lstm_model.eval()
    tokens = simple_tokenize(prefix)
    encoded = encode_prefix(tokens, seq_word_to_ix, max_len_seq)
    x = torch.tensor([encoded], dtype=torch.long).to(DEVICE)
    with torch.no_grad():
        logits = lstm_model(x)
        probs = torch.softmax(logits, dim=1)
        topk = torch.topk(probs, k=top_k, dim=1)
    results = []
    for idx, prob in zip(topk.indices[0].cpu().tolist(), topk.values[0].cpu().tolist()):
        results.append((seq_ix_to_word.get(idx, "[UNK]"), prob))
    return results

print("\nNext-word prediction examples:")
for prompt in [
    "the athlete",
    "the athlete trains in the",
    "the student studies in the",
    "the musician performs in the",
    "the researcher reads in the",
]:
    preds = predict_next_word(prompt, top_k=3)
    print(f"\nInput: '{prompt}'")
    for word, prob in preds:
        print(f"  {word:12s} prob={prob:.4f}")

plt.figure()
plt.plot(range(1, len(lstm_train_losses) + 1), lstm_train_losses, label="Train")
plt.plot(range(1, len(lstm_val_losses) + 1), lstm_val_losses, label="Validation")
plt.title("LSTM Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(range(1, len(lstm_train_accs) + 1), lstm_train_accs, label="Train Acc")
plt.plot(range(1, len(lstm_val_accs) + 1), lstm_val_accs, label="Validation Acc")
plt.plot(range(1, len(lstm_val_top3) + 1), lstm_val_top3, label="Validation Top-3 Acc")
plt.title("LSTM Accuracy Curves")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()


# ============================================================
# PART 3 - ATTENTION CLASSIFIER WITH TRAIN/VAL/TEST
# ============================================================

print("\n" + "=" * 70)
print("PART 3 - ATTENTION-BASED TEXT CLASSIFICATION")
print("=" * 70)

label_map = {"sports": 0, "education": 1, "arts": 2}
inv_label_map = {v: k for k, v in label_map.items()}

texts = [simple_tokenize(t) for t, label in labeled_sentences]
labels = [label_map[label] for _, label in labeled_sentences]

att_vocab = sorted(set(w for sent in texts for w in sent))
att_word_to_ix = {w: i + 1 for i, w in enumerate(att_vocab)}  # 0 = padding
att_vocab_size = len(att_word_to_ix) + 1
max_len_att = max(len(s) for s in texts)

def encode_text(words):
    ids = [att_word_to_ix[w] for w in words]
    ids += [0] * (max_len_att - len(ids))
    return ids

X_att = np.array([encode_text(t) for t in texts], dtype=np.int64)
y_att = np.array(labels, dtype=np.int64)

X_temp, X_test, y_temp, y_test = train_test_split(
    X_att, y_att, test_size=0.15, random_state=SEED, stratify=y_att
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.15, random_state=SEED, stratify=y_temp
)

X_train_t = torch.tensor(X_train, dtype=torch.long).to(DEVICE)
X_val_t = torch.tensor(X_val, dtype=torch.long).to(DEVICE)
X_test_t = torch.tensor(X_test, dtype=torch.long).to(DEVICE)

y_train_t = torch.tensor(y_train, dtype=torch.long).to(DEVICE)
y_val_t = torch.tensor(y_val, dtype=torch.long).to(DEVICE)
y_test_t = torch.tensor(y_test, dtype=torch.long).to(DEVICE)

class AttentionClassifier(nn.Module):
    def __init__(self, vocab_size, emb_dim=48, hidden_dim=64, num_classes=3):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.lstm = nn.LSTM(emb_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.attn = nn.Linear(hidden_dim * 2, 1)
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(hidden_dim * 2, num_classes)

    def forward(self, x):
        emb = self.embedding(x)
        outputs, _ = self.lstm(emb)
        attn_scores = self.attn(outputs).squeeze(-1)
        attn_weights = torch.softmax(attn_scores, dim=1)
        context = torch.sum(outputs * attn_weights.unsqueeze(-1), dim=1)
        logits = self.fc(self.dropout(context))
        return logits, attn_weights

att_model = AttentionClassifier(att_vocab_size).to(DEVICE)
att_loss_fn = nn.CrossEntropyLoss()
att_optimizer = optim.Adam(att_model.parameters(), lr=0.003)

att_train_losses, att_val_losses = [], []
att_train_accs, att_val_accs = [], []

for epoch in range(18):
    att_model.train()
    att_optimizer.zero_grad()
    train_logits, _ = att_model(X_train_t)
    train_loss = att_loss_fn(train_logits, y_train_t)
    train_loss.backward()
    nn.utils.clip_grad_norm_(att_model.parameters(), 1.0)
    att_optimizer.step()

    att_model.eval()
    with torch.no_grad():
        val_logits, _ = att_model(X_val_t)
        val_loss = att_loss_fn(val_logits, y_val_t)

        train_preds = torch.argmax(train_logits, dim=1)
        val_preds = torch.argmax(val_logits, dim=1)

        train_acc = (train_preds == y_train_t).float().mean().item()
        val_acc = (val_preds == y_val_t).float().mean().item()

    att_train_losses.append(train_loss.item())
    att_val_losses.append(val_loss.item())
    att_train_accs.append(train_acc)
    att_val_accs.append(val_acc)

    if (epoch + 1) % 3 == 0:
        print(
            f"Epoch {epoch + 1:2d} | Train Loss = {train_loss.item():.4f} | Val Loss = {val_loss.item():.4f} | "
            f"Train Acc = {train_acc:.4f} | Val Acc = {val_acc:.4f}"
        )

att_model.eval()
with torch.no_grad():
    test_logits, _ = att_model(X_test_t)
    test_preds = torch.argmax(test_logits, dim=1).cpu().numpy()

test_acc = accuracy_score(y_test, test_preds)
print(f"\nAttention classifier test accuracy: {test_acc:.4f}")

plt.figure()
plt.plot(range(1, len(att_train_losses) + 1), att_train_losses, label="Train")
plt.plot(range(1, len(att_val_losses) + 1), att_val_losses, label="Validation")
plt.title("Attention Classifier Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(range(1, len(att_train_accs) + 1), att_train_accs, label="Train")
plt.plot(range(1, len(att_val_accs) + 1), att_val_accs, label="Validation")
plt.title("Attention Classifier Training vs Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

cm = confusion_matrix(y_test, test_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["sports", "education", "arts"])
disp.plot()
plt.title("Attention Classifier Confusion Matrix")
plt.show()

def show_attention(sentence):
    tokens = simple_tokenize(sentence)
    encoded = encode_text(tokens)
    x = torch.tensor([encoded], dtype=torch.long).to(DEVICE)

    att_model.eval()
    with torch.no_grad():
        logits, weights = att_model(x)
        pred = torch.argmax(logits, dim=1).item()

    weights = weights[0][:len(tokens)].detach().cpu().numpy()

    print(f"\nSentence: {sentence}")
    print(f"Predicted class: {inv_label_map[pred]}")
    print("Attention weights:")
    for token, weight in zip(tokens, weights):
        print(f"  {token:12s} -> {weight:.4f}")

    plt.figure()
    plt.bar(tokens, weights)
    plt.title(f"Attention Weights: {sentence}")
    plt.xlabel("Token")
    plt.ylabel("Attention Weight")
    plt.xticks(rotation=30)
    plt.show()

show_attention("the dedicated athlete trains in the gym every morning")
show_attention("the focused student studies in the library every night")
show_attention("the creative musician performs in the studio every evening")


# ============================================================
# PART 4 - TRANSFER LEARNING WITH DISTILBERT
# Custom PyTorch loop, no datasets/trainer/accelerate needed
# ============================================================

print("\n" + "=" * 70)
print("PART 4 - TRANSFER LEARNING WITH DISTILBERT")
print("=" * 70)

positive_texts = [
    "I loved this movie and the acting was excellent",
    "The film was wonderful and very moving",
    "An amazing story with strong performances",
    "I really enjoyed the movie from start to finish",
    "This was a great film with a satisfying ending",
    "The performance was brilliant and emotional",
    "I liked the characters and the story a lot",
    "The movie was fun, warm, and entertaining",
    "An excellent film with beautiful direction",
    "The acting and soundtrack were fantastic",
    "I had a great time watching this movie",
    "The film was engaging and well written",
    "A smart and enjoyable story",
    "This movie was charming and heartfelt",
    "The cast did a wonderful job",
    "The plot was strong and the ending was great",
    "A very good movie that I would recommend",
    "This film was impressive and emotionally rich",
    "I was pleasantly surprised by how good it was",
    "The movie was thoughtful, moving, and strong",
    "I loved the visuals and the performances",
    "A creative and satisfying film",
    "The story was meaningful and beautifully told",
    "It was funny, touching, and very well made",
    "I enjoyed this film much more than I expected",
    "The script was sharp and the acting was excellent",
    "This was a powerful and inspiring movie",
    "A delightful film with excellent pacing",
    "The movie felt fresh, smart, and enjoyable",
    "I would happily watch this film again",
    "The experience was positive and memorable",
    "This was one of the better films I have seen lately",
    "The performances were strong and convincing",
    "The movie had heart and a strong message",
    "I genuinely liked this movie a lot",
    "The film was polished and very entertaining",
]

negative_texts = [
    "I hated this movie and the acting was awful",
    "The film was boring and badly written",
    "A disappointing story with weak performances",
    "I did not enjoy the movie at all",
    "This was a terrible film with a weak ending",
    "The performance was flat and unconvincing",
    "I disliked the characters and the story",
    "The movie was dull, slow, and forgettable",
    "A bad film with poor direction",
    "The acting and soundtrack were disappointing",
    "I wasted my time watching this movie",
    "The film was messy and poorly written",
    "A weak and frustrating story",
    "This movie was lifeless and predictable",
    "The cast did a poor job",
    "The plot was weak and the ending was bad",
    "A very bad movie that I would not recommend",
    "This film was underwhelming and emotionally empty",
    "I was surprised by how bad it was",
    "The movie was clumsy, slow, and weak",
    "I hated the visuals and the performances",
    "A confused and unsatisfying film",
    "The story was shallow and badly told",
    "It was annoying, flat, and poorly made",
    "I regret watching this film",
    "The script was weak and the acting was poor",
    "This was a frustrating and uninspired movie",
    "A forgettable film with terrible pacing",
    "The movie felt tired, obvious, and dull",
    "I would not watch this film again",
    "The experience was negative and disappointing",
    "This was one of the weaker films I have seen lately",
    "The performances were weak and unbelievable",
    "The movie had no heart and no clear message",
    "I genuinely disliked this movie a lot",
    "The film was sloppy and not entertaining",
]

all_texts = positive_texts + negative_texts
all_labels = [1] * len(positive_texts) + [0] * len(negative_texts)

# split into train/val/test
X_temp, X_test, y_temp, y_test = train_test_split(
    all_texts, all_labels, test_size=0.2, random_state=SEED, stratify=all_labels
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.2, random_state=SEED, stratify=y_temp
)

class TextClassificationDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=64):
        self.encodings = tokenizer(
            texts,
            truncation=True,
            padding=True,
            max_length=max_length,
            return_tensors="pt",
        )
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item["labels"] = self.labels[idx]
        return item

print("Loading tokenizer and pretrained model...")
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bert_model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2).to(DEVICE)

# freeze lower layers for speed/stability on laptop; unfreeze top transformer block + classifier head
for param in bert_model.distilbert.parameters():
    param.requires_grad = False

for param in bert_model.distilbert.transformer.layer[-1].parameters():
    param.requires_grad = True

for param in bert_model.pre_classifier.parameters():
    param.requires_grad = True
for param in bert_model.classifier.parameters():
    param.requires_grad = True

train_ds = TextClassificationDataset(X_train, y_train, tokenizer)
val_ds = TextClassificationDataset(X_val, y_val, tokenizer)
test_ds = TextClassificationDataset(X_test, y_test, tokenizer)

train_loader = DataLoader(train_ds, batch_size=8, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=8, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=8, shuffle=False)

bert_optimizer = optim.AdamW(filter(lambda p: p.requires_grad, bert_model.parameters()), lr=2e-5)

def evaluate_bert(model, loader):
    model.eval()
    losses = []
    preds = []
    trues = []
    with torch.no_grad():
        for batch in loader:
            labels = batch["labels"].cpu().numpy().tolist()
            batch = {k: v.to(DEVICE) for k, v in batch.items()}
            outputs = model(**batch)
            losses.append(outputs.loss.item())
            batch_preds = torch.argmax(outputs.logits, dim=1).cpu().numpy().tolist()
            preds.extend(batch_preds)
            trues.extend(labels)

    metrics = compute_classification_metrics(trues, preds)
    metrics["loss"] = float(np.mean(losses)) if losses else float("nan")
    return metrics, trues, preds

majority_baseline = max(Counter(y_train).values()) / len(y_train)

def keyword_baseline_predict(text):
    positive_keywords = {"love", "loved", "great", "excellent", "wonderful", "amazing", "enjoyed", "good", "satisfying", "fun"}
    negative_keywords = {"hate", "hated", "bad", "awful", "boring", "terrible", "weak", "poor", "dull", "disappointing"}
    toks = set(simple_tokenize(text))
    pos_hits = len(toks & positive_keywords)
    neg_hits = len(toks & negative_keywords)
    return 1 if pos_hits >= neg_hits else 0

keyword_preds = [keyword_baseline_predict(t) for t in X_test]
keyword_metrics = compute_classification_metrics(y_test, keyword_preds)

print(f"Majority baseline accuracy on training distribution: {majority_baseline:.4f}")
print(
    f"Keyword baseline test metrics -> "
    f"accuracy={keyword_metrics['accuracy']:.4f}, "
    f"precision={keyword_metrics['precision']:.4f}, "
    f"recall={keyword_metrics['recall']:.4f}, "
    f"f1={keyword_metrics['f1']:.4f}"
)

bert_train_losses = []
bert_val_losses = []
bert_val_accs = []

best_state = None
best_val_f1 = -1.0

for epoch in range(3):
    bert_model.train()
    running_loss = 0.0

    for batch in train_loader:
        batch = {k: v.to(DEVICE) for k, v in batch.items()}
        bert_optimizer.zero_grad()
        outputs = bert_model(**batch)
        loss = outputs.loss
        loss.backward()
        nn.utils.clip_grad_norm_(filter(lambda p: p.requires_grad, bert_model.parameters()), 1.0)
        bert_optimizer.step()
        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)
    val_metrics, _, _ = evaluate_bert(bert_model, val_loader)

    bert_train_losses.append(train_loss)
    bert_val_losses.append(val_metrics["loss"])
    bert_val_accs.append(val_metrics["accuracy"])

    print(
        f"Epoch {epoch + 1:2d} | "
        f"Train Loss = {train_loss:.4f} | Val Loss = {val_metrics['loss']:.4f} | "
        f"Val Acc = {val_metrics['accuracy']:.4f} | Val F1 = {val_metrics['f1']:.4f}"
    )

    if val_metrics["f1"] > best_val_f1:
        best_val_f1 = val_metrics["f1"]
        best_state = {k: v.cpu().clone() for k, v in bert_model.state_dict().items()}

if best_state is not None:
    bert_model.load_state_dict(best_state)

test_metrics, test_true, test_preds = evaluate_bert(bert_model, test_loader)

print("\nDistilBERT evaluation results:")
for k in ["accuracy", "precision", "recall", "f1", "loss"]:
    print(f"{k}: {test_metrics[k]:.4f}")

def predict_sentiment(text):
    bert_model.eval()
    encoded = tokenizer(text, truncation=True, padding=True, max_length=64, return_tensors="pt")
    encoded = {k: v.to(DEVICE) for k, v in encoded.items()}
    with torch.no_grad():
        outputs = bert_model(**encoded)
        pred = torch.argmax(outputs.logits, dim=1).item()
        prob = torch.softmax(outputs.logits, dim=1)[0, pred].item()
    return ("positive" if pred == 1 else "negative"), prob

examples = [
    "I absolutely loved the movie and the performances were excellent",
    "The film was a waste of time and painfully boring",
    "The story was strong, emotional, and beautifully acted",
    "I do not recommend this movie because it was dull and weak",
]

print("\nTransfer learning prediction examples:")
for ex in examples:
    label, prob = predict_sentiment(ex)
    print(f"'{ex}' -> {label} (confidence={prob:.4f})")

# Find and print a few errors
print("\nSample DistilBERT errors on the test set:")
shown = 0
for text, true_label in zip(X_test, y_test):
    pred_label, prob = predict_sentiment(text)
    pred_num = 1 if pred_label == "positive" else 0
    if pred_num != true_label and shown < 5:
        print(f"  Text: {text}")
        print(f"  True: {'positive' if true_label == 1 else 'negative'} | Pred: {pred_label} | confidence={prob:.4f}")
        shown += 1
if shown == 0:
    print("  No test errors found in this run.")

plt.figure()
plt.plot(range(1, len(bert_train_losses) + 1), bert_train_losses, label="Train")
plt.plot(range(1, len(bert_val_losses) + 1), bert_val_losses, label="Validation")
plt.title("DistilBERT Fine-Tuning Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(range(1, len(bert_val_accs) + 1), bert_val_accs)
plt.title("DistilBERT Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.ylim(0, 1.05)
plt.show()

metric_names = ["accuracy", "precision", "recall", "f1"]
metric_values = [test_metrics[m] for m in metric_names]

plt.figure()
plt.bar(metric_names, metric_values)
plt.title("DistilBERT Final Evaluation Metrics")
plt.xlabel("Metric")
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.show()

cm = confusion_matrix(test_true, test_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["negative", "positive"])
disp.plot()
plt.title("DistilBERT Confusion Matrix")
plt.show()


# ============================================================
# PART 5 - SYSTEM ANALYSIS
# ============================================================

print("\n" + "=" * 70)
print("PART 5 - SYSTEM ANALYSIS")
print("=" * 70)

print(f"""
What this improved program now does:

1. Word Representations
   - Trains CBOW embeddings on a much larger synthetic corpus
   - Evaluates with cosine similarity
   - Visualizes embedding structure with PCA
   - Tracks both training and validation loss

2. Sequence Modeling
   - Trains an LSTM for next-word prediction
   - Uses train/validation split
   - Reports top-1 and top-3 accuracy
   - Demonstrates realistic next-word predictions

3. Attention
   - Trains a BiLSTM + attention classifier
   - Uses train/validation/test split
   - Reports test accuracy and confusion matrix
   - Visualizes token-level attention weights

4. Transfer Learning
   - Fine-tunes pretrained DistilBERT using a custom PyTorch loop
   - Avoids datasets/Trainer/accelerate dependency issues
   - Uses train/validation/test split
   - Compares against simple baselines
   - Reports accuracy, precision, recall, F1, and confusion matrix
   - Shows sample prediction errors for error analysis

5. Why this is stronger than the earlier version
   - More data for learning
   - Proper validation and test evaluation
   - Baselines for comparison
   - Better diagnostics and more honest analysis
   - Still light enough to run on a MacBook Air

Done.
""")
```
