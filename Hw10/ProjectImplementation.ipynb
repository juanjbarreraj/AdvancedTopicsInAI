```python
# ============================================================
# MODULE 10 — Neural NLP System for Campus Agent Queries
#
# NOTE FOR JUPYTER USERS:
# Run this in a notebook cell. The plots are both saved to files
# and displayed inline so they appear in notebook output and in
# HTML exports from .ipynb.
#
# This module satisfies:
# Part 1: Word Representations (24.1)
# Part 2: Sequence Modeling (24.2–24.3)
# Part 3: Transformer or Attention (24.4)
# Part 4: Transfer Learning (24.5)
# Part 5: System Analysis (24.6)
# ============================================================

# If you are in Jupyter, this helps make plots show inline
try:
    get_ipython().run_line_magic("matplotlib", "inline")
except Exception:
    pass

import re
import random
from collections import Counter
from typing import List, Tuple, Dict, Any

import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

from IPython.display import display, Image as IPyImage

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# transformers is required for Part 4
try:
    from transformers import (
        AutoTokenizer,
        AutoModelForSequenceClassification,
        get_linear_schedule_with_warmup,
    )
    TRANSFORMERS_AVAILABLE = True
except ImportError:
    TRANSFORMERS_AVAILABLE = False

# ============================================================
# 0. Reproducibility and device
# ============================================================

RANDOM_SEED = 7
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ============================================================
# 1. Canonical schema
# ============================================================

TIME_SLOTS = [
    "07:00-08:00",
    "08:00-09:30",
    "09:40-11:10",
    "11:20-12:50",
    "13:00-14:30",
    "15:00-15:30",
    "16:00-17:30",
    "18:00-21:00"
]

TIME_ALIASES = {
    "07:00-08:00": ["7", "7am", "7 am", "07:00", "07:00-08:00", "early morning"],
    "08:00-09:30": ["8", "8am", "8 am", "08:00", "08:00-09:30", "this morning"],
    "09:40-11:10": ["9:40", "940", "09:40", "09:40-11:10", "mid morning"],
    "11:20-12:50": ["11:20", "11:20-12:50", "late morning"],
    "13:00-14:30": ["1", "1pm", "1 pm", "13:00", "13:00-14:30", "early afternoon"],
    "15:00-15:30": ["3", "3pm", "3 pm", "15:00", "15:00-15:30", "mid afternoon"],
    "16:00-17:30": ["4", "4pm", "4 pm", "16:00", "16:00-17:30", "late afternoon"],
    "18:00-21:00": ["6", "6pm", "6 pm", "18:00", "18:00-21:00", "tonight", "this evening"],
}

MAIN_LOCATIONS = [
    "LawrenceHall",
    "ThayerHall",
    "WestPenn",
    "StudentCenter",
    "Library",
    "VillagePark",
    "Track",
    "PlayHouse",
    "BoulevardStudios",
    "BoulevardAppartments",
    "OffCampus"
]

LOCATION_ALIASES = {
    "LawrenceHall": ["lawrence hall", "lawrencehall"],
    "ThayerHall": ["thayer hall", "thayerhall"],
    "WestPenn": ["west penn", "westpenn"],
    "StudentCenter": ["student center", "studentcenter", "the center", "center"],
    "Library": ["library", "the library"],
    "VillagePark": ["village park", "villagepark", "the park"],
    "Track": ["track", "the track"],
    "PlayHouse": ["playhouse", "play house"],
    "BoulevardStudios": ["boulevard studios", "boulevardstudios"],
    "BoulevardAppartments": ["boulevard appartments", "boulevard apartments", "boulevardappartments"],
    "OffCampus": ["off campus", "offcampus", "away from campus"],
}

SPORTS = ["TrackAndField", "Baseball"]
SPORT_ALIASES = {
    "TrackAndField": ["track and field", "trackandfield", "runner", "sprinter", "track athlete"],
    "Baseball": ["baseball", "ballplayer", "baseball player"]
}

STUDENT_TYPES = ["Athlete", "Copa", "Regular Student"]
STUDENT_TYPE_ALIASES = {
    "Athlete": ["athlete", "athletes", "student athlete"],
    "Copa": ["copa", "performing arts student", "conservatory student"],
    "Regular Student": ["regular student", "regular students", "student", "students"]
}

INTENT_LABELS = [
    "next_destination",
    "destination_reason",
    "current_location_reason",
    "next_location_prediction",
    "validity_check",
    "probability_to_location",
]
LABEL2ID = {label: i for i, label in enumerate(INTENT_LABELS)}
ID2LABEL = {i: label for label, i in LABEL2ID.items()}

# ============================================================
# 2. Text normalization
# ============================================================

def normalize_text(text: str) -> str:
    text = text.lower().strip()
    text = re.sub(r"[^\w\s:?-]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text

def simple_tokenize(text: str) -> List[str]:
    return normalize_text(text).split()

# ============================================================
# 3. Harder synthetic dataset generation
# ============================================================

FILLER_PREFIXES = [
    "",
    "hey ",
    "quick question ",
    "can you tell me ",
    "please tell me ",
    "i want to know ",
    "could you tell me ",
]

FILLER_SUFFIXES = [
    "",
    "?",
    " please",
    " right now",
    " if possible",
]

def sample_location_name() -> str:
    canonical = random.choice(MAIN_LOCATIONS)
    return random.choice(LOCATION_ALIASES[canonical])

def sample_time_phrase() -> str:
    slot = random.choice(TIME_SLOTS)
    return random.choice(TIME_ALIASES[slot])

def sample_student_type_phrase() -> str:
    t = random.choice(STUDENT_TYPES)
    return random.choice(STUDENT_TYPE_ALIASES[t])

def sample_sport_phrase() -> str:
    s = random.choice(SPORTS)
    return random.choice(SPORT_ALIASES[s])

def maybe_prefix(text: str) -> str:
    return random.choice(FILLER_PREFIXES) + text

def maybe_suffix(text: str) -> str:
    return text + random.choice(FILLER_SUFFIXES)

def maybe_noise(text: str) -> str:
    if random.random() < 0.15:
        text = text.replace("you", "u")
    if random.random() < 0.10:
        text = text.replace("going", "headed")
    if random.random() < 0.10:
        text = text.replace("probability", "chance")
    if random.random() < 0.08:
        text = text.capitalize()
    return text

def build_standard_templates() -> Dict[str, List[str]]:
    return {
        "next_destination": [
            "where are you going",
            "where are you going next",
            "where will you go next",
            "what is your next destination",
            "where is the {student_type} going",
            "where is the {sport} student going next",
            "are you going to {location} next",
            "where are you heading after {time}",
            "where do you go after this",
        ],
        "destination_reason": [
            "why are you going there",
            "why are you headed there",
            "why are you going to {location}",
            "why is the {student_type} going there",
            "why is the {sport} student going to {location}",
            "what is the reason you are going there",
            "why will you go to {location} next",
            "why are you moving toward {location}",
        ],
        "current_location_reason": [
            "why are you here",
            "why are you at {location}",
            "why are you in {location}",
            "why is the {student_type} here",
            "why is the {sport} student at {location}",
            "why are you still in {location}",
            "what explains being at {location} now",
        ],
        "next_location_prediction": [
            "where are you likely to be next",
            "where will you be next",
            "will you be in {location} next",
            "are you likely to be in {location} next",
            "what is your most likely next location",
            "where will the {student_type} be next",
            "where will the {sport} student be at {time}",
        ],
        "validity_check": [
            "are you in the right place right now",
            "are you supposed to be here",
            "are you supposed to be at {location}",
            "are you where you should be",
            "should the {student_type} be in {location} now",
            "is the {sport} student in the right place",
            "should you be here at {time}",
        ],
        "probability_to_location": [
            "what is the probability you go to {location}",
            "what is the probability you go to {location} next",
            "what are the chances you go to {location}",
            "what is the probability the {student_type} goes to {location}",
            "what is the probability the {sport} student goes to {location}",
            "how likely are you to be in {location} next",
            "estimate the probability you move to {location}",
        ],
    }

def build_hard_templates() -> Dict[str, List[str]]:
    return {
        "next_destination": [
            "after this where do you end up",
            "what place do you head toward next",
            "where would you most likely go from here",
            "what location comes next for you",
        ],
        "destination_reason": [
            "what makes that next place the right move",
            "why would you head over there",
            "what explains your move toward that place",
            "why is that your next stop",
        ],
        "current_location_reason": [
            "what explains your being here right now",
            "why does it make sense that you are here",
            "what puts you in this place at the moment",
            "why is this your current spot",
        ],
        "next_location_prediction": [
            "if i checked again in a minute where would you probably be",
            "what place are you most likely to be in next",
            "is {location} your likely next stop",
            "where do you probably end up next",
        ],
        "validity_check": [
            "does this location make sense for you right now",
            "are you currently where you ought to be",
            "is this the correct place for you at {time}",
            "should you actually be somewhere else right now",
        ],
        "probability_to_location": [
            "give me the chance that you move to {location}",
            "how probable is {location} as your next place",
            "what chance is there that you end up in {location}",
            "how likely is a move to {location} next",
        ],
    }

def generate_from_templates(
    templates: Dict[str, List[str]],
    n_per_intent: int
) -> Tuple[List[str], List[str]]:
    texts = []
    labels = []

    for intent, pattern_list in templates.items():
        for _ in range(n_per_intent):
            pattern = random.choice(pattern_list)
            text = pattern.format(
                location=sample_location_name(),
                time=sample_time_phrase(),
                student_type=sample_student_type_phrase(),
                sport=sample_sport_phrase()
            )
            text = maybe_prefix(text)
            text = maybe_suffix(text)
            text = maybe_noise(text)
            texts.append(normalize_text(text))
            labels.append(intent)

    return texts, labels

# ============================================================
# 4. Vocabulary and encoding
# ============================================================

PAD_TOKEN = "<pad>"
UNK_TOKEN = "<unk>"

def build_vocab(texts: List[str], min_freq: int = 1) -> Dict[str, int]:
    counter = Counter()
    for text in texts:
        counter.update(simple_tokenize(text))

    vocab = {PAD_TOKEN: 0, UNK_TOKEN: 1}
    for token, freq in counter.items():
        if freq >= min_freq:
            vocab[token] = len(vocab)
    return vocab

def encode_tokens(tokens: List[str], vocab: Dict[str, int], max_len: int) -> Tuple[List[int], int]:
    ids = [vocab.get(tok, vocab[UNK_TOKEN]) for tok in tokens][:max_len]
    length = len(ids)
    if len(ids) < max_len:
        ids += [vocab[PAD_TOKEN]] * (max_len - len(ids))
    return ids, length

# ============================================================
# 5. Datasets
# ============================================================

class IntentDataset(Dataset):
    def __init__(self, texts: List[str], labels: List[str], vocab: Dict[str, int], max_len: int = 24):
        self.texts = texts
        self.labels = [LABEL2ID[label] for label in labels]
        self.vocab = vocab
        self.max_len = max_len

        self.encoded = []
        self.lengths = []

        for text in texts:
            ids, length = encode_tokens(simple_tokenize(text), vocab, max_len)
            self.encoded.append(ids)
            self.lengths.append(length)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return {
            "input_ids": torch.tensor(self.encoded[idx], dtype=torch.long),
            "length": torch.tensor(self.lengths[idx], dtype=torch.long),
            "label": torch.tensor(self.labels[idx], dtype=torch.long),
            "text": self.texts[idx]
        }

class BERTIntentDataset(Dataset):
    def __init__(self, texts: List[str], labels: List[str], tokenizer, max_len: int = 40):
        self.texts = texts
        self.labels = [LABEL2ID[label] for label in labels]
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding="max_length",
            max_length=self.max_len,
            return_tensors="pt"
        )
        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long),
            "text": self.texts[idx]
        }

# ============================================================
# 6. Models
# ============================================================

class GRUIntentClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_classes, pad_idx):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=pad_idx)
        self.gru = nn.GRU(embed_dim, hidden_dim, batch_first=True)
        self.classifier = nn.Linear(hidden_dim, num_classes)

    def forward(self, input_ids, lengths):
        embedded = self.embedding(input_ids)
        packed = nn.utils.rnn.pack_padded_sequence(
            embedded, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        _, h_n = self.gru(packed)
        logits = self.classifier(h_n[-1])
        return logits

class AttentionBiGRUIntentClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_classes, pad_idx):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=pad_idx)
        self.bigru = nn.GRU(embed_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.attn = nn.Linear(hidden_dim * 2, 1)
        self.classifier = nn.Linear(hidden_dim * 2, num_classes)

    def forward(self, input_ids, lengths, return_attention=False):
        embedded = self.embedding(input_ids)
        packed = nn.utils.rnn.pack_padded_sequence(
            embedded, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        packed_out, _ = self.bigru(packed)
        outputs, _ = nn.utils.rnn.pad_packed_sequence(packed_out, batch_first=True)

        batch_size, seq_len, _ = outputs.size()
        mask = torch.arange(seq_len, device=lengths.device).unsqueeze(0) < lengths.unsqueeze(1)

        attn_scores = self.attn(outputs).squeeze(-1)
        attn_scores = attn_scores.masked_fill(~mask, -1e9)
        attn_weights = torch.softmax(attn_scores, dim=1)

        context = torch.bmm(attn_weights.unsqueeze(1), outputs).squeeze(1)
        logits = self.classifier(context)

        if return_attention:
            return logits, attn_weights
        return logits

# ============================================================
# 7. Plotting helpers (SAVE + SHOW INLINE)
# ============================================================

def show_saved_image(filename: str):
    try:
        display(IPyImage(filename=filename))
    except Exception:
        pass

def plot_training_history(history, title, filename):
    epochs = list(range(1, len(history["train_loss"]) + 1))

    plt.figure(figsize=(8, 5))
    plt.plot(epochs, history["train_loss"], label="Train loss")
    plt.plot(epochs, history["val_loss"], label="Val loss")
    plt.plot(epochs, history["val_f1"], label="Val macro-F1")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
    show_saved_image(filename)

def save_confusion_matrix(y_true, y_pred, labels, title, filename):
    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(labels))))

    plt.figure(figsize=(9, 7))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=45, ha="right")
    plt.yticks(tick_marks, labels)

    thresh = cm.max() / 2.0 if cm.max() > 0 else 1
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(
                j, i, format(cm[i, j], "d"),
                ha="center",
                va="center",
                color="white" if cm[i, j] > thresh else "black"
            )

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
    show_saved_image(filename)

def save_attention_plot(token_weights, filename):
    tokens = [t for t, _ in token_weights]
    weights = [w for _, w in token_weights]

    plt.figure(figsize=(10, 4))
    plt.bar(range(len(tokens)), weights)
    plt.xticks(range(len(tokens)), tokens, rotation=45, ha="right")
    plt.title("Attention Weights for Sample Query")
    plt.xlabel("Token")
    plt.ylabel("Attention weight")
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
    show_saved_image(filename)

# ============================================================
# 8. Training and evaluation
# ============================================================

def train_epoch_rnn(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0

    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        lengths = batch["length"].to(device)
        labels = batch["label"].to(device)

        optimizer.zero_grad()
        logits = model(input_ids, lengths)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * input_ids.size(0)

    return total_loss / len(loader.dataset)

@torch.no_grad()
def evaluate_rnn(model, loader, criterion, device):
    model.eval()
    total_loss = 0.0
    preds, golds, texts = [], [], []

    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        lengths = batch["length"].to(device)
        labels = batch["label"].to(device)

        logits = model(input_ids, lengths)
        loss = criterion(logits, labels)

        total_loss += loss.item() * input_ids.size(0)
        preds.extend(torch.argmax(logits, dim=1).cpu().tolist())
        golds.extend(labels.cpu().tolist())
        texts.extend(batch["text"])

    return {
        "loss": total_loss / len(loader.dataset),
        "accuracy": accuracy_score(golds, preds),
        "macro_f1": f1_score(golds, preds, average="macro"),
        "preds": preds,
        "golds": golds,
        "texts": texts
    }

def train_model_rnn(model, train_loader, val_loader, epochs, lr, device, model_name):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()

    history = {"train_loss": [], "val_loss": [], "val_acc": [], "val_f1": []}
    best_state = None
    best_val_f1 = -1.0

    for epoch in range(1, epochs + 1):
        train_loss = train_epoch_rnn(model, train_loader, optimizer, criterion, device)
        val_metrics = evaluate_rnn(model, val_loader, criterion, device)

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_metrics["loss"])
        history["val_acc"].append(val_metrics["accuracy"])
        history["val_f1"].append(val_metrics["macro_f1"])

        print(
            f"{model_name} | Epoch {epoch:02d} | "
            f"train_loss={train_loss:.4f} | "
            f"val_loss={val_metrics['loss']:.4f} | "
            f"val_acc={val_metrics['accuracy']:.4f} | "
            f"val_f1={val_metrics['macro_f1']:.4f}"
        )

        if val_metrics["macro_f1"] > best_val_f1:
            best_val_f1 = val_metrics["macro_f1"]
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

    if best_state is not None:
        model.load_state_dict(best_state)

    return history

# ============================================================
# 9. Embedding analysis
# ============================================================

def cosine_similarity(a, b):
    denom = (np.linalg.norm(a) * np.linalg.norm(b)) + 1e-12
    return float(np.dot(a, b) / denom)

def nearest_neighbors(model, vocab, query_word, top_k=5):
    if query_word not in vocab:
        return []

    emb_matrix = model.embedding.weight.detach().cpu().numpy()
    idx = vocab[query_word]
    query_vec = emb_matrix[idx]

    neighbors = []
    for word, word_idx in vocab.items():
        if word in {PAD_TOKEN, UNK_TOKEN, query_word}:
            continue
        sim = cosine_similarity(query_vec, emb_matrix[word_idx])
        neighbors.append((word, sim))

    neighbors.sort(key=lambda x: x[1], reverse=True)
    return neighbors[:top_k]

def analyze_embeddings(model, vocab):
    print("\n" + "=" * 70)
    print("PART 1 — WORD REPRESENTATIONS / EMBEDDING SIMILARITY")
    print("=" * 70)

    probe_words = ["track", "library", "student", "center", "probability", "going", "here", "next"]
    for word in probe_words:
        if word in vocab:
            nbrs = nearest_neighbors(model, vocab, word, top_k=5)
            print(f"\nNearest neighbors for '{word}':")
            for nbr, sim in nbrs:
                print(f"  {nbr:20s} cosine={sim:.4f}")
        else:
            print(f"\nWord '{word}' not present in vocab.")

# ============================================================
# 10. Attention analysis
# ============================================================

@torch.no_grad()
def inspect_attention(model, text, vocab, max_len=24):
    model.eval()
    tokens = simple_tokenize(text)
    input_ids, length = encode_tokens(tokens, vocab, max_len)

    x = torch.tensor([input_ids], dtype=torch.long, device=DEVICE)
    l = torch.tensor([length], dtype=torch.long, device=DEVICE)

    logits, attn = model(x, l, return_attention=True)
    pred = torch.argmax(logits, dim=1).item()
    attn = attn.squeeze(0).cpu().numpy()[:length]
    return ID2LABEL[pred], list(zip(tokens, attn.tolist()))

# ============================================================
# 11. Transfer learning
# ============================================================

def train_epoch_bert(model, loader, optimizer, scheduler, device):
    model.train()
    total_loss = 0.0

    for batch in loader:
        optimizer.zero_grad()

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()

        total_loss += loss.item() * input_ids.size(0)

    return total_loss / len(loader.dataset)

@torch.no_grad()
def evaluate_bert(model, loader, device):
    model.eval()
    preds = []
    golds = []
    total_loss = 0.0

    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

        logits = outputs.logits
        loss = outputs.loss

        total_loss += loss.item() * input_ids.size(0)
        preds.extend(torch.argmax(logits, dim=1).cpu().tolist())
        golds.extend(labels.cpu().tolist())

    return {
        "loss": total_loss / len(loader.dataset),
        "accuracy": accuracy_score(golds, preds),
        "macro_f1": f1_score(golds, preds, average="macro"),
        "preds": preds,
        "golds": golds
    }

def run_transfer_learning(train_texts, train_labels, val_texts, val_labels, test_texts, test_labels):
    if not TRANSFORMERS_AVAILABLE:
        raise ImportError("transformers is not installed.")

    print("\n" + "=" * 70)
    print("PART 4 — TRANSFER LEARNING WITH DISTILBERT")
    print("=" * 70)

    model_name = "distilbert-base-uncased"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=len(INTENT_LABELS),
        id2label=ID2LABEL,
        label2id=LABEL2ID
    ).to(DEVICE)

    train_ds = BERTIntentDataset(train_texts, train_labels, tokenizer, max_len=40)
    val_ds = BERTIntentDataset(val_texts, val_labels, tokenizer, max_len=40)
    test_ds = BERTIntentDataset(test_texts, test_labels, tokenizer, max_len=40)

    train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=16)
    test_loader = DataLoader(test_ds, batch_size=16)

    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
    total_steps = 2 * len(train_loader)
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=max(1, total_steps // 10),
        num_training_steps=total_steps
    )

    history = {"train_loss": [], "val_loss": [], "val_acc": [], "val_f1": []}
    best_state = None
    best_val_f1 = -1.0

    for epoch in range(1, 3):
        train_loss = train_epoch_bert(model, train_loader, optimizer, scheduler, DEVICE)
        val_metrics = evaluate_bert(model, val_loader, DEVICE)

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_metrics["loss"])
        history["val_acc"].append(val_metrics["accuracy"])
        history["val_f1"].append(val_metrics["macro_f1"])

        print(
            f"DistilBERT | Epoch {epoch:02d} | "
            f"train_loss={train_loss:.4f} | "
            f"val_loss={val_metrics['loss']:.4f} | "
            f"val_acc={val_metrics['accuracy']:.4f} | "
            f"val_f1={val_metrics['macro_f1']:.4f}"
        )

        if val_metrics["macro_f1"] > best_val_f1:
            best_val_f1 = val_metrics["macro_f1"]
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

    if best_state is not None:
        model.load_state_dict(best_state)

    test_metrics = evaluate_bert(model, test_loader, DEVICE)

    plot_training_history(
        history,
        title="DistilBERT Fine-Tuning History",
        filename="module10_distilbert_training.png"
    )
    save_confusion_matrix(
        test_metrics["golds"],
        test_metrics["preds"],
        INTENT_LABELS,
        title="DistilBERT Confusion Matrix",
        filename="module10_distilbert_confusion.png"
    )

    print("\nDistilBERT test performance:")
    print(f"  Accuracy : {test_metrics['accuracy']:.4f}")
    print(f"  Macro-F1 : {test_metrics['macro_f1']:.4f}")

    return model, tokenizer, test_metrics

# ============================================================
# 12. Main
# ============================================================

def main():
    print("\n" + "=" * 70)
    print("MODULE 10 — NEURAL NLP SYSTEM FOR CAMPUS AGENT QUERIES")
    print("=" * 70)
    print(f"Using device: {DEVICE}")

    # Standard dataset
    standard_texts, standard_labels = generate_from_templates(build_standard_templates(), n_per_intent=220)

    # Harder eval set
    hard_texts, hard_labels = generate_from_templates(build_hard_templates(), n_per_intent=50)

    print(f"\nStandard dataset size: {len(standard_texts)}")
    print("Standard intent distribution:")
    print(Counter(standard_labels))

    print(f"\nHard evaluation set size: {len(hard_texts)}")
    print("Hard-set intent distribution:")
    print(Counter(hard_labels))

    train_texts, temp_texts, train_labels, temp_labels = train_test_split(
        standard_texts,
        standard_labels,
        test_size=0.30,
        random_state=RANDOM_SEED,
        stratify=standard_labels
    )

    val_texts, test_texts, val_labels, test_labels = train_test_split(
        temp_texts,
        temp_labels,
        test_size=0.50,
        random_state=RANDOM_SEED,
        stratify=temp_labels
    )

    print(f"\nTrain size: {len(train_texts)}")
    print(f"Val size  : {len(val_texts)}")
    print(f"Test size : {len(test_texts)}")

    vocab = build_vocab(train_texts, min_freq=1)
    max_len = 24

    train_ds = IntentDataset(train_texts, train_labels, vocab, max_len=max_len)
    val_ds = IntentDataset(val_texts, val_labels, vocab, max_len=max_len)
    test_ds = IntentDataset(test_texts, test_labels, vocab, max_len=max_len)
    hard_ds = IntentDataset(hard_texts, hard_labels, vocab, max_len=max_len)

    train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=64)
    test_loader = DataLoader(test_ds, batch_size=64)
    hard_loader = DataLoader(hard_ds, batch_size=64)

    # --------------------------------------------------------
    # Part 2 — GRU
    # --------------------------------------------------------
    print("\n" + "=" * 70)
    print("PART 2 — SEQUENCE MODELING WITH GRU")
    print("=" * 70)

    gru_model = GRUIntentClassifier(
        vocab_size=len(vocab),
        embed_dim=100,
        hidden_dim=128,
        num_classes=len(INTENT_LABELS),
        pad_idx=vocab[PAD_TOKEN]
    ).to(DEVICE)

    gru_history = train_model_rnn(
        gru_model, train_loader, val_loader, epochs=8, lr=1e-3, device=DEVICE, model_name="GRU"
    )

    criterion = nn.CrossEntropyLoss()
    gru_test = evaluate_rnn(gru_model, test_loader, criterion, DEVICE)
    gru_hard = evaluate_rnn(gru_model, hard_loader, criterion, DEVICE)

    plot_training_history(gru_history, "GRU Training History", "module10_gru_training.png")
    save_confusion_matrix(
        gru_test["golds"], gru_test["preds"], INTENT_LABELS,
        "GRU Confusion Matrix", "module10_gru_confusion.png"
    )

    print("\nGRU regular test performance:")
    print(f"  Accuracy : {gru_test['accuracy']:.4f}")
    print(f"  Macro-F1 : {gru_test['macro_f1']:.4f}")

    print("\nGRU hard test performance:")
    print(f"  Accuracy : {gru_hard['accuracy']:.4f}")
    print(f"  Macro-F1 : {gru_hard['macro_f1']:.4f}")

    # --------------------------------------------------------
    # Part 3 — Attention-BiGRU
    # --------------------------------------------------------
    print("\n" + "=" * 70)
    print("PART 3 — ATTENTION-BASED BI-GRU")
    print("=" * 70)

    attn_model = AttentionBiGRUIntentClassifier(
        vocab_size=len(vocab),
        embed_dim=100,
        hidden_dim=128,
        num_classes=len(INTENT_LABELS),
        pad_idx=vocab[PAD_TOKEN]
    ).to(DEVICE)

    attn_history = train_model_rnn(
        attn_model, train_loader, val_loader, epochs=8, lr=1e-3, device=DEVICE, model_name="Attention-BiGRU"
    )

    attn_test = evaluate_rnn(attn_model, test_loader, criterion, DEVICE)
    attn_hard = evaluate_rnn(attn_model, hard_loader, criterion, DEVICE)

    plot_training_history(attn_history, "Attention-BiGRU Training History", "module10_attention_gru_training.png")
    save_confusion_matrix(
        attn_test["golds"], attn_test["preds"], INTENT_LABELS,
        "Attention-BiGRU Confusion Matrix", "module10_attention_gru_confusion.png"
    )

    print("\nAttention-BiGRU regular test performance:")
    print(f"  Accuracy : {attn_test['accuracy']:.4f}")
    print(f"  Macro-F1 : {attn_test['macro_f1']:.4f}")

    print("\nAttention-BiGRU hard test performance:")
    print(f"  Accuracy : {attn_hard['accuracy']:.4f}")
    print(f"  Macro-F1 : {attn_hard['macro_f1']:.4f}")

    print("\nAttention effect analysis:")
    print(f"  GRU regular Macro-F1             : {gru_test['macro_f1']:.4f}")
    print(f"  Attention-BiGRU regular Macro-F1 : {attn_test['macro_f1']:.4f}")
    print(f"  GRU hard Macro-F1                : {gru_hard['macro_f1']:.4f}")
    print(f"  Attention-BiGRU hard Macro-F1    : {attn_hard['macro_f1']:.4f}")

    sample_query = "will you be in the center next"
    pred_label, token_weights = inspect_attention(attn_model, sample_query, vocab, max_len=max_len)
    save_attention_plot(token_weights, "module10_attention_weights.png")

    print(f"\nSample attention analysis for query: '{sample_query}'")
    print(f"Predicted intent: {pred_label}")
    for tok, wt in token_weights:
        print(f"  {tok:15s} weight={wt:.4f}")

    # --------------------------------------------------------
    # Part 1 — embeddings
    # --------------------------------------------------------
    analyze_embeddings(attn_model, vocab)

    # --------------------------------------------------------
    # Part 4 — Transfer learning
    # --------------------------------------------------------
    bert_test = None
    if TRANSFORMERS_AVAILABLE:
        _, _, bert_test = run_transfer_learning(
            train_texts, train_labels,
            val_texts, val_labels,
            test_texts, test_labels
        )
    else:
        print("\n" + "=" * 70)
        print("PART 4 — TRANSFER LEARNING")
        print("=" * 70)
        print("transformers is not installed, so DistilBERT fine-tuning was skipped.")
        print("Install with:")
        print("  pip install transformers")
        print("or on your Mac if needed:")
        print("  python3 -m pip install transformers --break-system-packages")

    # --------------------------------------------------------
    # Part 5 — System analysis
    # --------------------------------------------------------
    print("\n" + "=" * 70)
    print("PART 5 — SYSTEM ANALYSIS")
    print("=" * 70)

    print("\nModel comparison:")
    print(f"  GRU regular Accuracy              : {gru_test['accuracy']:.4f}")
    print(f"  GRU regular Macro-F1              : {gru_test['macro_f1']:.4f}")
    print(f"  GRU hard Accuracy                 : {gru_hard['accuracy']:.4f}")
    print(f"  GRU hard Macro-F1                 : {gru_hard['macro_f1']:.4f}")
    print(f"  Attention-BiGRU regular Accuracy  : {attn_test['accuracy']:.4f}")
    print(f"  Attention-BiGRU regular Macro-F1  : {attn_test['macro_f1']:.4f}")
    print(f"  Attention-BiGRU hard Accuracy     : {attn_hard['accuracy']:.4f}")
    print(f"  Attention-BiGRU hard Macro-F1     : {attn_hard['macro_f1']:.4f}")
    if bert_test is not None:
        print(f"  DistilBERT regular Accuracy       : {bert_test['accuracy']:.4f}")
        print(f"  DistilBERT regular Macro-F1       : {bert_test['macro_f1']:.4f}")

    print("\nDetailed classification report for Attention-BiGRU on regular test set:")
    print(classification_report(
        attn_test["golds"],
        attn_test["preds"],
        target_names=INTENT_LABELS,
        digits=4
    ))

    print("\nDetailed classification report for Attention-BiGRU on hard test set:")
    print(classification_report(
        attn_hard["golds"],
        attn_hard["preds"],
        target_names=INTENT_LABELS,
        digits=4
    ))

    print("Limitations and challenges:")
    print("- The dataset is still synthetic, even though it is now noisier and more varied.")
    print("- Very high scores on the standard split may still overestimate real-world performance.")
    print("- The hard test set is a better stress test, but it is still generated rather than collected from real users.")
    print("- Embedding neighbors can still look noisy because the corpus is relatively small.")
    print("- Attention weights show what the model emphasizes, but they are not a perfect explanation of reasoning.")
    print("- This module focuses on neural intent classification, not full semantic reasoning.")
    print("- The best future integration would combine this neural classifier with the semantic campus-agent pipeline from Module 9.")

    print("\nSaved files:")
    print("  module10_gru_training.png")
    print("  module10_gru_confusion.png")
    print("  module10_attention_gru_training.png")
    print("  module10_attention_gru_confusion.png")
    print("  module10_attention_weights.png")
    if bert_test is not None:
        print("  module10_distilbert_training.png")
        print("  module10_distilbert_confusion.png")
if __name__ == "__main__":
    main()
```

### Module 10 — Neural NLP System for Campus Agent Queries

This module extends the semester-long campus project by implementing a neural natural language system for classifying campus-agent questions. The underlying project idea remains the same: a user can ask questions such as “Where are you going?” or “Will you be in the center next?” and the system interprets the question type.

### Part 1 — Word representations
Text is represented using learned embeddings. The sequence models begin with an embedding layer that maps tokens into dense vectors. After training, the embedding space can be analyzed using cosine similarity to examine which words become semantically close in the campus-query domain.

### Part 2 — Sequence modeling
A GRU classifier is trained to predict the intent of a campus-agent question. This treats the query as a sequence of tokens and models token order, making it more expressive than a bag-of-words system.

### Part 3 — Attention
An attention-based bidirectional GRU is also trained. The attention mechanism assigns weights to tokens in a query and produces a weighted representation of the sequence. This allows analysis of which words matter most for prediction.

### Part 4 — Transfer learning
A pretrained DistilBERT model is fine-tuned on the same intent classification task. This is the transfer learning component. Instead of learning only from scratch on synthetic data, the model starts from a large pretrained language representation and adapts it to the campus-query task.

### Part 5 — System analysis
Performance is evaluated using:
- accuracy
- macro-F1
- confusion matrices
- classification reports

The system is evaluated both on a standard test split and on a harder paraphrased test set. This makes the evaluation more realistic than using only clean template-based examples.

### Limitations
The dataset remains synthetic, even though it has been made noisier and more varied. Therefore, very high performance on the standard split still likely overestimates real-world accuracy. The hard test set provides a more meaningful stress test, but it is still not the same as real user language. The module focuses on neural intent classification and would be most powerful when integrated with the semantic reasoning pipeline from Module 9.
