### Module 11 — Computer Vision (Justification for Exclusion)

Although Module 11 focuses on designing a computer vision system to interpret visual data, it was intentionally not integrated into the final project.

The primary objective of this project is to develop an interactive campus intelligence system that models student movement, decision-making, and behavior, and allows users to query agents through natural language. The system is centered around three core components:

* probabilistic and reinforcement learning models for agent behavior,
* natural language processing for user interaction,
* and an interactive frontend for visualization and exploration.

Integrating a computer vision system would introduce a separate modality (image-based perception) that does not directly enhance the core functionality of the project. The campus environment is already fully structured and digitally defined, meaning that agent positions, movements, and identities are known explicitly through the simulation. Therefore, applying image processing, feature extraction, or object detection would be redundant rather than necessary.

Additionally, implementing a meaningful computer vision component (such as object detection or 3D reasoning) would require building or sourcing a realistic visual dataset, which is outside the scope and timeline of this project. Including such a component would risk fragmenting the system and weakening the coherence of the overall design.

Given that Module 9 (NLP) and Module 10 (Neural Systems) already provide strong alignment with the project’s goals—enabling intelligent interaction and learning-based interpretation—the decision was made to prioritize depth and integration within these areas rather than adding an unrelated computer vision subsystem.

In conclusion, Module 11 was not included in the final project to maintain a focused, cohesive system that emphasizes intelligent agent behavior and natural language interaction, which are the central contributions of this work.


### Transition to Final Project Development

At this stage of the project, the focus shifts from implementing additional course modules to building a cohesive, interactive system for the final presentation. While Module 11 (Computer Vision) introduces valuable concepts, it does not directly enhance the core objective of this project, which is centered on modeling, simulating, and interacting with student behavior across a campus environment. Incorporating computer vision would introduce unnecessary complexity without meaningfully improving the system’s functionality or clarity.

Instead, the project is now evolving into a fully integrated application that combines the work from previous modules into a single, interactive experience. The goal is to demonstrate not only the underlying AI techniques but also how they function together in a realistic and user-facing system.

### Current Direction

The project is now being developed as a **live simulation platform** with the following characteristics:

* A visual representation of the campus using a structured map layout
* Multiple agents (students) moving across the campus
* Behavior driven by previously implemented models (probabilistic reasoning, decision-making, and reinforcement learning)
* A user interface that allows interaction with individual agents
* Natural language queries that connect to the NLP system developed in Module 9

### Key Components Being Built

1. **Campus Visualization**
   The campus is mapped onto a coordinate system aligned with a real layout. Locations are positioned deliberately to reflect spatial relationships, creating a more realistic and interpretable environment.

2. **Agent-Based Simulation**
   Multiple agents representing different student types (Athlete, Copa, Regular Student) are placed on the map. Each agent has:

   * A current location
   * A target destination
   * Behavior influenced by time, type, and learned policies

3. **Time-Driven System**
   The simulation introduces a dynamic clock:

   * Starts at 8:00 AM
   * Runs at an accelerated rate (5× real time)
   * Drives agent movement and decision-making
   * Allows manual time-slot selection for demonstration purposes

4. **Animated Movement**
   Agents do not “teleport” between locations. Instead, their movement is visualized through smooth transitions across the map, making behavior easier to interpret and more engaging to observe.

5. **Interactive Interface**
   Users can:

   * Click on any agent
   * View detailed state information (type, location, destination, etc.)
   * Ask questions using natural language

6. **NLP Integration**
   The system connects to the Module 9 NLP pipeline, enabling responses such as:

   * Predicting where an agent is going
   * Explaining why a movement decision was made
   * Providing probabilistic or reasoning-based answers

### Why This Direction

This approach transforms the project from a collection of independent assignments into a unified system that demonstrates:

* Probabilistic reasoning (Modules 1–2)
* Decision-making under uncertainty (Module 2–3)
* Sequential planning and reinforcement learning (Module 3–8)
* Natural language understanding and response (Module 9)
* Neural modeling and attention mechanisms (Module 10)

By focusing on integration and interaction rather than adding unrelated features, the project becomes clearer, more impactful, and more aligned with the original goal.

### Final Objective

The final deliverable is a **live, interactive demonstration** where a user can:

* Observe agents moving across a campus in real time
* Select specific agents and inspect their behavior
* Ask questions about their decisions and receive meaningful responses
* Understand how multiple AI techniques combine within a single system

This represents a transition from **implementing algorithms** to **building an intelligent system**, which is the primary goal of the final project.
